```python
import pandas as pd
import random

# Sample DataFrame
df=pd.read_csv("jd.csv")
```

```python
df["label"]=1
```

```python
df=df.dropna(axis=1)
```

```python
df
```

```python

# Number of negative samples to generate per job
num_negatives_per_job = 2

# Create a list to store negative samples
negative_samples = []

# Generate negative samples
for _, row in df.iterrows():
    job = row['title']
    description = row['desc']
    
    # Generate 'num_negatives_per_job' negative samples
    for _ in range(num_negatives_per_job):
        # Randomly select a description that is not the current job's description
        while True:
            random_index = random.randint(0, len(df) - 1)
            random_description = df.loc[random_index, 'desc']
            if random_description != description:
                break
        
        negative_samples.append({
            'title': job,
            'desc': random_description,
            'label': 0  # Negative label
        })

# Convert negative samples to a DataFrame
negative_df = pd.DataFrame(negative_samples)

# Combine the original DataFrame and the negative DataFrame
combined_df = pd.concat([df, negative_df], ignore_index=True)

# print(combined_df)
```

```python
combined_df.to_csv("jd_pos_neg.csv",index=False)
```
